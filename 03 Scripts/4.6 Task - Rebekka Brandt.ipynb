## Combining orders_products_combined with product data

```python
# Drop '_merge' column for df_orders_products_combined
df_orders_prods_comb = df_orders_prods_comb.drop(columns = ['_merge'])
```

```python
# Drop 'Unnamend' column for df_orders_products_combined
df_prod = df_prod.drop(columns = ['Unnamed: 0'])
```

```python
# combining orders_products_combined with product data
df_merged = df_orders_prods_comb.merge(df_prod, on = 'product_id', indicator = True)
```
