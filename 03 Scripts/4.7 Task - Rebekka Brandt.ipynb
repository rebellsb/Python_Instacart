## If /Else

```python
# creating a function for 'price label' with if/else statement
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# executing function
df['price_range'] = df.apply(price_label, axis=1)
```

## Creating 'busiest and least busy days'

```python
# changing loop to include 2 busiest and 2 slowest days

result_busiest = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0 or value== 1:
    result_busiest.append("Busiest days")
  elif value == 3 or value ==4:
    result_busiest.append("Least busy days")
  else:
    result_busiest.append("Regularly busy days")
```

## Creating 'busiest period of the day'

```python
# checking for busiest hour of the day
df_ords_prods_merged['order_hour_of_day'].value_counts(dropna = False)
```

```python
# creating loop for busiest period of day

result_busiest_pod = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value in[10,11,14,15,13,12,16,9]:
    result_busiest_pod.append("Most Orders")
  elif value in[17, 8, 18,19,20,7,21,22]:
    result_busiest_pod.append("Average Orders")
  else:
    result_busiest_pod.append("Fewest Orders")
```

```python
# add busiest period of day as column
df_ords_prods_merged['busiest_period_of_day'] = result_busiest_pod
```
